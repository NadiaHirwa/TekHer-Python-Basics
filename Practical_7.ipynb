### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
values = [100, -20, 10, -30, 50]
cumulative = np.cumsum([0] + values[:-1])
plt.figure()
for i in range(len(values)):
    color = 'green' if values[i] >= 0 else 'red'
    plt.bar(i, values[i], bottom=cumulative[i], color=color)
plt.title("Waterfall Chart")
plt.show()
```
